```python
from collections import deque

def find_augmenting_path(graph, matched, start):
    queue = deque([start])
    visited = set([start])
    parent = {}

    while queue:
        current = queue.popleft()
        for neighbor in graph[current]:
            if neighbor not in visited:
                visited.add(neighbor)
                queue.append(neighbor)
                parent[neighbor] = current
                if neighbor not in matched or (neighbor in matched and parent[neighbor] not in matched):
                    # Found augmenting path
                    return parent, neighbor

    return None, None

def find_maximum_matching(n, edges):
    graph = {i: [] for i in range(1, n + 1)}
    for edge in edges:
        u, v = edge
        graph[u].append(v)
        graph[v].append(u)

    matched = {}
    for node in range(1, n + 1):
        parent, augmenting_end = find_augmenting_path(graph, matched, node)

        while augmenting_end is not None:
            current = parent[augmenting_end]
            matched[current] = augmenting_end
            matched[augmenting_end] = current

            parent, augmenting_end = find_augmenting_path(graph, matched, node)

    pairs = [(min(node, matched[node]), max(node, matched[node])) for node in matched if node < matched[node]]
    return pairs

input_str = '''8 13
1 2
1 4
1 6
1 8
3 4
3 6
3 8
5 6
5 8
7 8
7 6
7 4
7 2'''

input_list = input_str.split('\n')
n, m = map(int, input_list[0].split())
edges = [tuple(map(int, line.split())) for line in input_list[1:]]

pairs = find_maximum_matching(n, edges)
print(len(pairs))
for pair in pairs:
    print(pair[0], pair[1])
```
